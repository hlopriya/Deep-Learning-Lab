```python
import numpy as np
# for AND gate (Input/Output)
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 5
w,b = np.zeros(2),0

for _ in range(epochs):
  for x,y in zip(x_list,y_list):
    z = np.dot(x,w) + b
    ### condition area for y_pred
    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    print("error : ",E,"<- at epoch",_)
    w += E*x
    b += E
    print(w,b)
```

```python
# Plot Accuracy vs Iterations (Epoch-like behavior)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load dataset
df = pd.read_csv("/content/spam_or_not_spam.csv")
X_text = df["email"].fillna("")
y = df["label"]


X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# TF-IDF
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_train = vectorizer.fit_transform(X_train_text)
X_test = vectorizer.transform(X_test_text)

# Train model while tracking accuracy
accuracies = []
iterations = range(1, 21)

for i in iterations:
    model = LogisticRegression(max_iter=i, warm_start=True)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

# Plot
plt.figure()
plt.plot(iterations, accuracies)
plt.xlabel("Iterations")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Iterations")
plt.show()
```

```python
# =====================================================
# EMAIL CLASSIFICATION
# Multiple Learning Rates
# Accuracy Graph and Loss Graph (Separate)
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, log_loss

# -----------------------------
# 1. Load Dataset
# -----------------------------
df = pd.read_csv("/content/spam_or_not_spam.csv")

X_text = df["email"].fillna("")
y = df["label"]

# -----------------------------
# 2. Train-Test Split
# -----------------------------
X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# -----------------------------
# 3. TF-IDF Vectorization
# -----------------------------
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_train = vectorizer.fit_transform(X_train_text)
X_test = vectorizer.transform(X_test_text)

# -----------------------------
# 4. Multiple Learning Rates
# -----------------------------
learning_rates = [0.0001, 0.001, 0.01, 0.1]
epochs = 20

accuracy_results = {}
loss_results = {}

for lr in learning_rates:

    model = SGDClassifier(
        loss="log_loss",
        learning_rate="constant",
        eta0=lr,
        max_iter=1,
        warm_start=True,
        random_state=42
    )

    accuracies = []
    losses = []

    for epoch in range(epochs):
        model.fit(X_train, y_train)

        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)

        acc = accuracy_score(y_test, y_pred)
        loss = log_loss(y_test, y_prob)

        accuracies.append(acc)
        losses.append(loss)

    accuracy_results[lr] = accuracies
    loss_results[lr] = losses

# -----------------------------
# 5. Accuracy Graph
# -----------------------------
plt.figure()
for lr in learning_rates:
    plt.plot(range(1, epochs+1), accuracy_results[lr], label=f"LR={lr}")

plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Epochs for Different Learning Rates")
plt.legend()
plt.show()
print("\n")
# -----------------------------
# 6. Loss Graph
# -----------------------------
plt.figure()
for lr in learning_rates:
    plt.plot(range(1, epochs+1), loss_results[lr], label=f"LR={lr}")

plt.xlabel("Epochs")
plt.ylabel("Log Loss")
plt.title("Loss vs Epochs for Different Learning Rates")
plt.legend()
plt.show()
```
